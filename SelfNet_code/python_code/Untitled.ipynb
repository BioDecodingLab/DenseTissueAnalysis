```python
import os
import numpy as np
import tifffile
from scipy.ndimage import label

# Carpeta que contiene los _masks.tif
input_dir = r"\\WS1\WS1_Remote_Disk\Current Segovia Lab\Emilio Gutiérrez\Emilio Gutiérrez SOS\test Cellpose\train set sin bordes\rebanadas 2D"

# Recorrer todos los archivos
for filename in os.listdir(input_dir):
    if filename.endswith("_masks.tif"):
        mask_path = os.path.join(input_dir, filename)
        seg_path = os.path.join(input_dir, filename.replace("_masks.tif", "_seg.npy"))

        # Leer la máscara
        mask = tifffile.imread(mask_path)

        # Convertir a binario si contiene valores como 255
        if mask.max() > 1:
            mask = (mask > 0).astype(np.uint8)

        # Etiquetar objetos conectados
        mask_labeled, num_objects = label(mask)

        if num_objects == 0:
            print(f"⚠️  ADVERTENCIA: {filename} no tiene objetos segmentados (solo fondo). No se guardó.")
            continue  # Saltar si no hay objetos

        # Guardar como diccionario que Cellpose entiende
        np.save(seg_path, {'masks': mask_labeled})
        print(f"✅ Guardado: {os.path.basename(seg_path)} con {num_objects} objetos.")
```

```python
import os
import numpy as np
import tifffile

# Carpeta de entrada con los archivos .npy
input_dir = r"\\WS1\WS1_Remote_Disk\Current Segovia Lab\Emilio Gutiérrez\Emilio Gutiérrez SOS\test Cellpose\test set\npy"

# Carpeta de salida para guardar los .tif
output_dir = r"\\WS1\WS1_Remote_Disk\Current Segovia Lab\Emilio Gutiérrez\Emilio Gutiérrez SOS\test Cellpose\test set\masks"

# Crear la carpeta de salida si no existe
os.makedirs(output_dir, exist_ok=True)

# Recorrer todos los archivos .npy en la carpeta de entrada
for filename in os.listdir(input_dir):
    if filename.endswith(".npy"):
        npy_path = os.path.join(input_dir, filename)
        base_name = os.path.splitext(filename)[0]
        tif_path = os.path.join(output_dir, base_name + "_masks.tif")

        try:
            data = np.load(npy_path, allow_pickle=True).item()
            if 'masks' in data:
                masks = data['masks']
                tifffile.imwrite(tif_path, masks.astype(np.uint16))
                print(f"✅ Guardado: {tif_path}")
            else:
                print(f"⚠️  {filename} no contiene la clave 'masks'")
        except Exception as e:
            print(f"❌ Error con {filename}: {e}")
```
